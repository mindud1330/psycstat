```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm, t
```

```python
x = np.linspace(-5, 5, 1000)
```

```python
df = 4
std_normal_pdf = norm.pdf(x)
t_pdf = t.pdf(x, df=df)
```

```python
alpha = 0.05
z_critical = norm.ppf(1 - alpha / 2)
t_critical = t.ppf(1 - alpha / 2, df)
t_critical_19 = t.ppf(1 - alpha / 2, 19)

print(f"Z-test 임계값: {z_critical}")
print(f"t-test 임계값 (df={df}): {t_critical}")
print(f"t-test 임계값 (df=19): {t_critical_19}")
```

```python
plt.figure(figsize=(12, 6))

# z 
plt.plot(x, std_normal_pdf, label='Standard Normal Distribution', color='blue')
plt.fill_between(x, std_normal_pdf, 
                 where=(x > z_critical) | (x < -z_critical), 
                 color='blue', alpha=0.3)

# t
plt.plot(x, t_pdf, label='t-Distribution (df=4)', color='red')
plt.fill_between(x, t_pdf, 
                 where=(x > t_critical) | (x < -t_critical), 
                 color='red', alpha=0.3)

# critical value line
plt.axvline(z_critical, color='blue', linestyle='dotted')
plt.axvline(-z_critical, color='blue', linestyle='dotted')
plt.axvline(t_critical, color='red', linestyle='dotted')
plt.axvline(-t_critical, color='red', linestyle='dotted')

plt.title('Comparison of Standard Normal Distribution and t-Distribution (df=4)')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()
```
